```python
import pandas as pd
```

```python
data = pd.read_csv('credit_customers.csv')
```

```python
data.head()
```

```python
# Display all column names in the dataframe
data.columns

# Show detailed information about the dataframe including data types and non-null values
data.info()

# Display the dimensions of the dataframe (rows, columns)
data.shape
```

```python
# Remove unnecessary columns from the dataset
# - checking_status: Account status information not needed for analysis
# - duration: Time duration feature being dropped
# - purpose: Purpose of credit not required for current analysis
# - personal_status: Personal status information being excluded
data.drop(columns=[
    'checking_status',
    'duration',
    'purpose',
    'personal_status'
], inplace=True)
```

```python
data.columns
```

```python
data.shape
```

```python
data
```

```python
# Convert column names to lowercase and replace spaces with underscores for consistent naming convention
data.columns = data.columns.str.lower().str.replace(' ', '_')
```

```python
data.columns
```

```python
# Check if there are any duplicate rows in the dataset
# Returns the count of duplicated rows (rows that appear more than once, with only subsequent occurrences counted as duplicates)
data.duplicated().sum()
```

```python
# remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# check nulls
data.isna().sum()
```

```python
# remove null rows
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.info()
```

```python
data
```

```python
data['employment'].unique()
```

```python
# Clean the 'employment' column by removing leading/trailing whitespace and converting to lowercase
data['employment'] = data['employment'].str.strip().str.lower()
```

```python
data['employment'].unique()
```

```python
# Create a mapping dictionary to convert employment categories to numeric values
# '<1' and 'unemployed' are mapped to 0 years
# '1<=x<4' is mapped to 2 years (midpoint)
# '4<=x<7' is mapped to 5 years (midpoint)
# '>=7' is mapped to 7 years (minimum value)
employment_mapping = {
    '<1': 0,
    '1<=x<4': 2,
    '4<=x<7': 5,
    '>=7': 7,
    'unemployed': 0
}

# Apply the mapping to convert the categorical 'employment' column to numeric values
data['employment'] = data['employment'].map(employment_mapping)
```

```python
data['employment'].unique()
```

```python
data['employment'].isna().sum()
```

```python
data
```

```python
# Convert the 'class' column values to binary:
# - 'good' becomes 1
# - any other value becomes 0
data['class'] = data['class'].apply(lambda x: 1 if x == 'good' else 0)
```

```python
# Count the occurrences of each unique value in the 'class' column
# This helps us understand the distribution of classes in the dataset
data['class'].value_counts()
```

```python
data.dtypes
```

```python
data
```

```python
# Define a list of numerical features to be used in the analysis
# - age: customer's age
# - existing_credits: number of existing credits at this bank
# - num_dependents: number of people being financially supported by the customer
# - residence_since: years at current residence
cols = ['age', 'existing_credits', 'num_dependents', 'residence_since']
```

```python
data[cols].isna().sum()
```

```python
# Round all values in the specified columns to the nearest integer
data[cols] = data[cols].round()
```

```python
# Convert specified columns to integer data type
# This ensures numerical operations work correctly and removes any decimal points
data[cols] = data[cols].astype(int)
```
